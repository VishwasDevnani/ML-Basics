```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Creating Dataset

```python
def create_data(size, m):
    x = np.random.uniform(1, size, (size,1))
    random = np.random.random((size,1))
    y = m*x+10*random
    print(np.mean(random))
    return x,y
```

```python
x,y=create_data(100,2)
```

### Calculate y_hat

```python
def y_(x, m ,c):
    return m*x + c
```

```python
y_(x,5,1).shape
```

## MSE

```python
def mse(y, y_):
    return np.mean(np.square(y-y_))
```

## Derivatives

```python
def del_m(y, y_, x):
    return -2 * np.dot((y-y_).T, x)[0][0]/y.shape[0]
```

```python
def del_c(y, y_):
    return -2 * np.mean(y-y_)
```

# Gradient Descent

```python
def gd(x, y, alpha, eps):
    i = 0
    m,c = -5, 1
    y_hat = y_(x,m,c)
    print(y_hat.shape)
    loss = mse(y,y_hat)
    print(loss.shape)
    
    while True:
        print(i,loss)
        m = m-alpha*del_m(y,y_hat,x)
        c = c-alpha*del_c(y,y_hat)
        
        y_hat = y_(x,m,c)
        new_loss = mse(y,y_hat)
        
        if abs(loss-new_loss)<eps:
            return m,c
        loss = new_loss
        i+=1
```

```python
gd(x,y,0.0001, 0.0000001)
```

```python
m,c = (2.0060497792868883, 4.535562816880193)
```

```python
y_hat = y_(x,m,c)
```

```python
plt.scatter(x,y)
plt.plot(x, y_hat, color='red')
plt.show()
```

```python
def std_x(x):
    return (x-np.mean(x))/np.std(x)
```

```python
x_std = std_x(x)
```

```python
gd(x_std,y,0.01, 0.0000001)
```
