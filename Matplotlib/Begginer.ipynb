```python
x = np.array([2,3,6,9,15])
y = np.array([1,5,4,7,1])
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'--')
```

```python
plt.plot(x,y,'o--b')
```

```python
plt.plot(x,y,'o--g')
```

```python
plt.plot(x,y,'o--y',mec='g',markersize=8,linestyle='dotted')
```

```python
plt.plot(x,y,'o--y',mec='g',markersize=18,linestyle='dotted',mfc = 'orange')
```

```python
plt.plot(x,y,'o--y',mec='g',markersize=18,linestyle='dotted',mfc = 'orange',linewidth = '4')
```

```python
plt.plot(x,y,linewidth = 5,label ='2022')
plt.plot(x+5,y+10,linewidth=5, color = 'green', label ='2023')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('TechBrain', loc = 'left', color='Red')
plt.grid(axis='y',linestyle='dashed',linewidth = 5,color='pink')
plt.legend()
```

```python
plt.figure(figsize=(8,5))
plt.subplot(2,2,1)
plt.plot(x,y)
plt.xlabel('x1')
plt.ylabel('y1')
plt.subplot(2,2,3)
plt.plot(x+np.random.randint(0,3,5),y+np.random.randint(0,3,5),color='green')
plt.subplot(2,2,4)
plt.plot(x+np.random.randint(1,5,5),y+np.random.randint(0,3,5),color='red')
plt.subplot(2,2,2)
plt.plot(x+np.random.randint(0,3,5),y+np.random.randint(0,6,5),color='yellow')
```
